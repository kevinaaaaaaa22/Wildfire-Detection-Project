```python
from matplotlib import path
from PIL import Image

count = 0

pdFires = pd.read_pickle("C:\\Users\\yangz\\Desktop\\FireProject\\pdFires.pkl") # Load historical wildfire data

for i in range (0, 4531):
    polyAcres = pdFires.at[i, "poly_Acres"]
    if polyAcres < 40: # Skip fires that burnt less than 40 acres
        print ("Skipped " + str(i))
        continue
    
    filePath = "C:\\Users\\yangz\\Desktop\\FireProjectBatchDownloadNew\\Perimeter\\" + str(count) + "_" + str(pdFires.at[i, "poly_Incid"]) + "_" + str(i) + ".npy"

    try:
        with open(filePath, 'rb') as f1:
            testTemp = np.load(f1)
            
        count += 1
            
        perimeter = pdFires.at[i, "geometry"] # Perimeter polygon

        bounds = list(perimeter.bounds)
        bounds[0] = bounds[0]-0.005
        bounds[1] = bounds[1]-0.005
        bounds[2] = bounds[2]+0.005
        bounds[3] = bounds[3]+0.005

        xLeft = bounds[0]
        yTop = bounds[3]
        xDist = abs(bounds[2]-bounds[0])
        yDist = abs(bounds[3]-bounds[1])
        xDistNew = 444
        yDistNew = 340
        exteriorList = list(perimeter.exterior.coords)
        xNewList = []
        yNewList = []
        polygonList = []
        polygonListSP = []
        for each in exteriorList:
            x = each[0]
            y = each[1]
            xNew = ((x-xLeft)/xDist) * xDistNew
            yNew = ((yTop-y)/yDist) * yDistNew
            xNewList.append(xNew)
            yNewList.append(yNew)
            polygonList.append([xNew, yNew])
            polygonListSP.append((xNew, yNew))
        p = path.Path(polygonListSP)

        label = np.zeros((340, 444))

        for a in range (0, 340):
            for b in range (0, 444):
                if p.contains_points([(b, a)])[0] == True: # If point is part of perimeter polygon, shade dark
                    label[a][b] = 200
                else:
                    label[a][b] = 0

        label = label.astype("uint8")
        stackedImg = np.stack((label, )*3, axis=-1)

        im = Image.fromarray(stackedImg) # Turn into images for training BASNet model
        im.save("C:\\Users\\yangz\\Desktop\\FireProjectBatchDownloadNew\\PerimeterLabels\\Label" + str(count) + ".png")

        print ("Processed " + str(i))

    except:
        print ("Skipped " + str(i))
        continue
```
