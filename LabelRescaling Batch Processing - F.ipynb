```python
import numpy as np
import pandas as pd
from PIL import Image
import cv2
import matplotlib.pyplot as plt
```

```python
for i in range (1, 1061):
    try:
        im = cv2.imread("C:\\Users\\yangz\\Desktop\\FireProjectBatchDownloadNew\\PerimeterLabels\\Label" + str(i) + ".png")
        imMin = np.min(im)
        imMax = np.max(im)
        im = (im-imMin)/(imMax-imMin)
        im = im*255

        im = im.astype("uint8")
        im2 = Image.fromarray(im)
        im2 = im2.convert("RGB")
        im2.save("C:\\Users\\yangz\\Desktop\\FireProjectBatchDownloadNew\\PerimeterLabelsRescaledV4\\LabelR-V4-" + str(i) + ".png")

        print ("Processed " + str(i))
    except:
        print ("Skipped " + str(i))
```

```python
imTest = cv2.imread("C:\\Users\\yangz\\Desktop\\FireProjectBatchDownloadNew\\PerimeterLabelsRescaledV3\\LabelR-V3-117.png")
imTest = imTest.astype("uint8")
```

```python
def plot_image(image, factor=1.0, clip_range = None, **kwargs):
    """
    Utility function for plotting RGB images.
    """
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(15, 15))
    if clip_range is not None:
        pos = ax.imshow(np.clip(image * factor, *clip_range), **kwargs)
        fig.colorbar(pos, ax=ax)
        
    else:
        ax.imshow(image * factor, **kwargs)
    ax.set_xticks([])
    ax.set_yticks([])
```

```python
plot_image(imTest)
```
