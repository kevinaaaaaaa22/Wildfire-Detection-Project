```python
import numpy as np
import pandas as pd
from PIL import Image
import cv2
import matplotlib.pyplot as plt
```

```python
for i in range (0, 1060):
    try:
        im = cv2.imread("C:\\Users\\yangz\\Desktop\\FireProjectBatchDownloadNew\\PerimeterPreProcessed\\PerimeterProcessed" + str(i) + ".png")
        imMin = np.min(im)
        imMax = np.max(im)
        im = (im-imMin)/(imMax-imMin)
        im = im*155
        im = im+100
        im[0][0] = 0

        im = im.astype("uint8")
        im2 = Image.fromarray(im)
        im2 = im2.convert("RGB")
        im2.save("C:\\Users\\yangz\\Desktop\\FireProjectBatchDownloadNew\\PerimeterPreProcessedRescaled\\PerimeterProcessedR" + str(i) + ".jpg")

        print ("Processed " + str(i))
    except:
        print ("Skipped " + str(i))
```
